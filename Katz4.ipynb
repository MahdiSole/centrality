```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

A1 = np.matrix([[0, 1, 0, 0], 
               [1, 0, 1, 0],
               [0, 1, 0, 1],
               [0, 0, 1, 0],])

G1 = nx.from_numpy_matrix(A1)

pos = {0: (4, 0.3), 1: (4, 0), 2: (-1, 0.3), 3: (-1, 0)}


nx.draw_networkx(G1, pos)


ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()

eigenvaluesG1 = np.linalg.eigvals(A1)
print (eigenvaluesG1)
print (max (eigenvaluesG1))
rG1 = 1/(max (eigenvaluesG1))
print (rG1)



centralityG11 = nx.katz_centrality(G1, 0.1)

centralityG12 = nx.katz_centrality(G1, 0.2)

centralityG13 = nx.katz_centrality(G1, 0.35)

centralityG14 = nx.katz_centrality(G1, 0.4)

centralityG15 = nx.katz_centrality(G1, 0.6)



fig, ax = plt.subplots()
ax.plot(range(len(centralityG11)), list(centralityG11.values()), label = '0.1')
plt.xticks(np.arange(min(range(len(centralityG11))), max(range(len(centralityG11)))+1, 1.0))

ax.plot(range(len(centralityG13)), list(centralityG13.values()), label = '0.35')
plt.xticks(np.arange(min(range(len(centralityG11))), max(range(len(centralityG11)))+1, 1.0))

ax.plot(range(len(centralityG15)), list(centralityG15.values()), label = '0.6')
plt.xticks(np.arange(min(range(len(centralityG11))), max(range(len(centralityG11)))+1, 1.0))

ax.set(xlabel='Node', ylabel='Katz Centrality')

ax.legend()
plt.show()
```
